# Python'da For Döngüleri

Python'da for döngüleri, belirli bir kod bloğunu belirli bir kez ya da bir dizi/veri yapısının üzerinde gezinerek tekrar tekrar çalıştırmak için kullanılır. İşte bu konuda bilmen gereken bazı temel ve kritik noktalar:

## 1. Temel Sözdizimi
```python
for item in iterable:
    # Yapılacak işlemler
```
- `iterable`, üzerinde gezinilebilecek herhangi bir nesnedir (liste, demet, sözlük, dize vb.).
- `item`, `iterable` içindeki her bir elemanı temsil eder.

## 2. İterable Nesneler
For döngüleri, listeler, demetler, sözlükler, dizeler ve hatta dosyalar gibi iterable nesnelerle kullanılabilir.

## 3. Range Fonksiyonu
```python
for i in range(10):
    print(i)
```
Bu kod 0'dan 9'a kadar olan sayıları yazdırır.

## 4. Sözlüklerle Kullanım
```python
for key, value in my_dict.items():
    print(key, value)
```

## 5. İç İçe Döngüler
Bir döngü içinde başka bir döngü tanımlayabilirsin.

## 6. Enumerate Fonksiyonu
```python
for index, value in enumerate(my_list):
    print(index, value)
```

## 7. List Comprehensions
```python
squares = [x**2 for x in range(10)]
```

## 8. Break ve Continue
- `break`, döngüyü tamamen sonlandırmak için kullanılır.
- `continue`, döngünün mevcut iterasyonunu atlayıp bir sonraki iterasyona geçmek için kullanılır.

## 9. Else Bloğu
Python'daki for döngülerinde, döngü normal şekilde (yani `break` ile sonlandırılmadan) sona erdiğinde çalışacak bir `else` bloğu tanımlayabilirsin.
