# <font color='green'> <b>Importing Libraries </b><font color='black'>

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

# <font color='green'> <b>Missing Values</b><font color='black'>

```python
id_no = ["P001", "P002", "P003", "P004", "P005", "P006", "P007", "P008", "P009", "P010", "P011"]
gender = ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F", "M"]
status = ["FT", "PT", "-", "FT", "PT", "PT", "FT", "-", "PT", "FT", np.nan]
dept = ["DS", "FS", "AWS", "AWS", "DS", None, "FS", "FS", np.nan, "DS", "AWS"]
V1 = np.array([2, 3, 5, np.nan, 7, 1, np.nan, 10, 14, "-", 6])
V2 = np.array([8, np.nan, 5, 8, 11, np.nan, np.nan, 2, 3, 7, 9])
salary = np.array([np.nan, 54, 59, 120, 58, 75, None, 136, 60, 125, np.nan])

df0 = pd.DataFrame({
                   "id" : id_no,
                   "gender": gender,
                   "status": status,
                   "dept": dept,
                   "var1" : V1,
                   "var2" : V2,
                   "salary" : salary
                  })
```

```python
df = df0.copy()       # df0 ın orjinal olarak kalmasını istiyoruz
df
```

```python
df["var1"] = df["var1"].replace(to_replace = "-", value = np.nan).astype("float")

df["var1"]
```

```python
df["status"] = df["status"].replace("-", np.nan)
df["status"]
```

```python
df.isnull().sum()
```

## <font color='blue'> <b>Filling Missing Values (Imputation)</b><font color='black'>

```python
df
```

```python
df.info()
```

### <font color='orange'> <b>Filling Numeric Variables</b><font color='orange'>

```python
df.fillna(0) # bütün nan değerleri sıfır ile doldurmak istersek bu şekilde kullanılabilir. 
```

```python
df
```

```python
df.var1.fillna(0) # pd.Series iel de fillna yapılabilir. 
```

```python
df.var1.mean()
```

```python
df.var1.fillna(df.var1.mean()) # df.var1 in ortalaması ile df var1 in nan değerlerini doldurduk
```

```python
df.var1.sort_values()
```

```python
df.var1.median()
```

```python
df.var1.fillna(df.var1.median()) # df.var1 in ortanca(median) değeri ile df var1 in nan değerlerini doldurduk
```

```python
df
```

### <font color='orange'> <b>Filling by condition & group</b><font color='black'>

```python
df
```

```python
df.groupby("status")["salary"].mean()
```

```python
df.loc[df.salary > 90, "status" ]
```

```python
df.loc[df.salary > 90, "status" ].mode()[0]
```

```python
df.loc[df.salary <= 90, "status" ]
```

```python
df.loc[df.salary <= 90, "status" ].mode()[0]
```

```python
df.loc[df.salary > 90, "status" ].fillna(df.loc[df.salary > 90, "status" ].mode()[0])
```

```python
# inplace= True çalışmadı. 
df.loc[df.salary > 90, "status" ].fillna(df.loc[df.salary > 90, "status" ].mode()[0], inplace=True)
```

```python
df
```

```python
#inplace çalışamzsa assign etmek gerekir. 
df.loc[df.salary > 90, "status" ] = df.loc[df.salary > 90, "status" ].mode()[0]
```

```python
df
```

```python
df.loc[df.salary <= 90, "status" ] = df.loc[df.salary <= 90, "status" ].mode()[0]
```

```python
df
```

**'Status' sütunundaki son eksik değeri 'gender' ve 'dept' grubunun modu ile dolduralım.**

```python
df.groupby(["gender", "dept"])["status"].transform(lambda x: x.mode()[0])
```

```python
df
```

```python
df.status.fillna(df.groupby(["gender", "dept"])["status"].transform(lambda x: x.mode()[0]))
```

```python
df
```

```python
df.var1.mode()
```

```python
df.var1.fillna(df.var1.mode())
```

```python
df.status.fillna(df.groupby(["gender", "dept"])["status"].transform(lambda x: x.mode()[0]), inplace=True)
```

```python
df
```

```python
df.var1.fillna(df.groupby(["gender", "status"])["var1"].transform("mean"), inplace=False)
```

```python
df.var1.fillna(df.groupby(["gender", "status"])["var1"].transform("mean"), inplace=True)
```

```python
df.var2.fillna(df.groupby("status")["var2"].transform("mean"), inplace=False)
```

```python
df.var2.fillna(df.groupby("status")["var2"].transform("mean"), inplace=True)
```

```python
df
```

```python
df.salary.fillna(df.groupby(["status","dept"])["salary"].transform("mean"), inplace=False)
```

```python
df.salary.fillna(df.groupby(["status","dept"])["salary"].transform("mean"), inplace=True)
```
