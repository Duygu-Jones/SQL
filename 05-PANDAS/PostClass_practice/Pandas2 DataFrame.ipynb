[pandas.com](https://pandas.pydata.org/docs/reference/index.html)

'cmap=', matplotlib kütüphanesinde kullanılan renk haritalarını (color map) belirtmek için kullanılan bir parametredir. 

-viridis    : Mavi-yeşil renk geçişleri olan bu renk haritası, sık kullanılan ve renk körü dostu bir seçenektir.
-plasma     : Magenta-sarı renk geçişlerini içeren bir renk haritasıdır, viridis gibi renk körleri için de uygun bir seçenektir.
-inferno    : Sarı-turuncu-kahverengi tonlarının kullanıldığı bir renk haritasıdır.
-magma      : Siyah-mor renk geçişleri olan bir renk haritasıdır.
-cividis    : Renk körlerine uygun, sarı-mavi tonlarını içeren bir renk haritasıdır.

```python
df.plot.scatter(x='age',y='fare', c='pclass', cmap ='viridis')  # cmap= color map
```
