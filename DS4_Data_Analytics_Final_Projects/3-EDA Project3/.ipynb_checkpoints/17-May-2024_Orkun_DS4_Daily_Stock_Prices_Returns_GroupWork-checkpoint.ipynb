```python
# Using Matplotlib, plot lineplots that display all 4 stocks daily prices on one single figure.
import pandas as pd
import matplotlib.pyplot as plt

# Reload the data
prices_df = pd.read_csv('stocks_daily_prices.csv')
prices_df['Date'] = pd.to_datetime(prices_df['Date'])
prices_df.set_index('Date', inplace=True)

# Plot the time series of stock prices on a single figure
plt.figure(figsize=(14, 7))
for column in prices_df.columns:
    plt.plot(prices_df.index, prices_df[column], label=column)
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Using Matplotlib, plot 4 stocks daily prices on multiple subplots.

# Create subplots for each stock's daily prices
fig, axs = plt.subplots(4, 1, figsize=(14, 14), sharex=True)

# Plot each stock's daily prices in separate subplots
for i, column in enumerate(prices_df.columns):
    axs[i].plot(prices_df.index, prices_df[column], label=column)
    axs[i].set_title(f'{column} Daily Prices')
    axs[i].set_ylabel('Price')
    axs[i].legend()
    axs[i].grid(True)

# Set the xlabel for the last subplot
axs[-1].set_xlabel('Date')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Using Matplotlib, plot a pie chart that shows asset allocations, using the 'explode' attribute to increase the separation between AAPL and the rest of the portfolio.

# Define the asset allocations
allocations = [40, 25, 20, 15]  # Example percentages for AAPL, JPM, PG, UAL
labels = ['AAPL', 'JPM', 'PG', 'UAL']

# Define the explode parameter to separate AAPL from the rest
explode = [0.1, 0, 0, 0]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(allocations, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Asset Allocations')
plt.show()
```

```python
# Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color. Display the mean and standard deviation for both stocks on top of the figure.

# Calculate the mean and standard deviation for UAL and PG returns
ual_mean = returns_df['UAL'].mean()
ual_std = returns_df['UAL'].std()
pg_mean = returns_df['PG'].mean()
pg_std = returns_df['PG'].std()

# Plot the histograms for UAL and PG returns
plt.figure(figsize=(14, 7))

# Histogram for UAL
plt.subplot(1, 2, 1)
plt.hist(returns_df['UAL'], bins=40, color='red', alpha=0.7)
plt.title(f'UAL Returns\nMean: {ual_mean:.2f}, Std: {ual_std:.2f}')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.grid(True)

# Histogram for PG
plt.subplot(1, 2, 2)
plt.hist(returns_df['PG'], bins=40, color='red', alpha=0.7)
plt.title(f'PG Returns\nMean: {pg_mean:.2f}, Std: {pg_std:.2f}')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D plot for daily returns of JPM, AAPL, and UAL
fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for JPM, AAPL, and UAL daily returns
ax.scatter(returns_df['JPM'], returns_df['AAPL'], returns_df['UAL'], c='b', marker='o')

# Set labels for axes
ax.set_xlabel('JPM Daily Returns')
ax.set_ylabel('AAPL Daily Returns')
ax.set_zlabel('UAL Daily Returns')
ax.set_title('3D Plot of Daily Returns: JPM, AAPL, and UAL')

plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Veri setini yeniden y√ºkleyelim
returns_df = pd.read_csv('stocks_daily_returns.csv')
returns_df['Date'] = pd.to_datetime(returns_df['Date'])
returns_df.set_index('Date', inplace=True)

# Create a DataFrame for plotting with plotly
plot_df = returns_df.reset_index()

# Create the figure
fig = go.Figure()

# Add JPM scatter plot
fig.add_trace(go.Scatter3d(
    x=plot_df['Date'],
    y=plot_df['JPM'],
    z=plot_df['UAL'],
    mode='markers',
    marker=dict(color='blue', size=5),
    text=plot_df['Date'].dt.strftime('%Y-%m-%d'),
    name='JPM'
))

# Add AAPL scatter plot
fig.add_trace(go.Scatter3d(
    x=plot_df['Date'],
    y=plot_df['AAPL'],
    z=plot_df['UAL'],
    mode='markers',
    marker=dict(color='red', size=5),
    text=plot_df['Date'].dt.strftime('%Y-%m-%d'),
    name='AAPL'
))

# Add UAL scatter plot
fig.add_trace(go.Scatter3d(
    x=plot_df['Date'],
    y=plot_df['AAPL'],
    z=plot_df['JPM'],
    mode='markers',
    marker=dict(color='green', size=5),
    text=plot_df['Date'].dt.strftime('%Y-%m-%d'),
    name='UAL'
))

# Set the titles and labels
fig.update_layout(
    title='3D Plot of Daily Returns: JPM, AAPL, and UAL',
    scene=dict(
        xaxis_title='Date',
        yaxis_title='Daily Returns',
        zaxis_title='Daily Returns'
    )
)

# Show the plot
fig.show()
```
