```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## MatPlotLib
### Line

```python
# Data Olusturma

x= np.linspace(0,10,20)
y= 2*x
```

```python
#Grafik olusturma

plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.show()
```

### Scatter Graph

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D", "E"])
y = np.array([5, 11, 3, 14, 8])
```

### Bar Chart

```python
plt.figure(figsize=(5,3))
plt.bar(x, y)
plt.show()
```

### Histogram Graph

```python
x= np.random.normal(170, 10, 250)  # (mean, std, obj)
```

```python
plt.figure(figsize=(5,3))
plt.hist(x)
```

### Box Plot

```python
x= np.random.normal(170, 10, 250)  # (mean, std, obj) # degerler random alindigi icin her cell calistirildiginda degisir
plt.figure(figsize=(5,3))
plt.boxplot(x)
```

### Pie Chart

```python
x = ['Kategori 1', 'Kategori 2', 'Kategori 3', 'Kategori 4']
```

```python
import matplotlib.pyplot as plt

# Veri
sizes = [15, 30, 45, 10]

# Pasta grafiini olutur
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=x)

# Gster
plt.axis('equal')
plt.show()
```
