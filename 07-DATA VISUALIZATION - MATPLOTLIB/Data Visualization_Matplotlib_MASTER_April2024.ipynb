# Grafiklere argüman verme

### Kod Açıklaması

Bu kod bloğunda, bir sinüs dalgası çizmek için Matplotlib kullanılmıştır.

1. `x = np.linspace(0, 10, 100)`: Bu komut, 0 ile 10 arasında eşit aralıklı 100 sayı üretir. Bu sayılar, x-eksenindeki değerlerimizi oluşturur.

2. `y = np.sin(x)`: Numpy'nin `sin` fonksiyonunu kullanarak, `x` dizisinin her bir değeri için sinüs değerlerini hesaplar. Bu değerler, y-eksenindeki değerlerimizi oluşturur.

3. `plt.figure(figsize=(8, 6))`: Grafiğin boyutunu 8x6 birim olarak ayarlar. Bu, grafik boyutunu özelleştirmek için kullanılır.

4. `plt.plot(x, y, "r")`: `x` ve `y` dizilerini kullanarak bir çizgi grafiği çizer ve çizgi rengini kırmızı ("r") olarak belirler.

5. `plt.show()`: Çizilen grafiği gösterir. Bu komut, grafiğin ekranda görünmesini sağlar.

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize = (8, 6))
plt.plot(x, y, "r")

plt.show()
````

#### Kod Çıktısı Yorumlaması

0 ile 10 arasındaki değerler için bir sinüs dalgası grafiği görülmektedir. Grafiğin boyutu, 8x6 birim olarak ayarlanmıştır, bu da detayların daha kolay görülmesini sağlar. Kırmızı renkli çizgi (`"r"` parametresi ile belirlenmiş), sinüs dalgasının dalgalanmalarını net bir şekilde gösterir. `np.linspace` kullanılarak oluşturulan x-ekseni değerleri, sinüs fonksiyonunun sürekli ve düzgün bir şekilde nasıl değiştiğini sergiler. Bu tür grafikler, matematiksel fonksiyonların görselleştirilmesinde veya periyodik verilerin analizinde sıkça kullanılır.

# Çoklu data ve legend verme - Plotting multiple sets of data

### Kod Açıklaması

Bu kod bloğu, yaşa göre maaş dağılımını iki farklı veri seti kullanarak göstermek için Matplotlib kullanılmıştır.

1. `age = np.arange(25, 44, 2.5)`: 25 yaşından başlayarak, 2.5'lik artışlarla 44 yaşına kadar olan yaşları bir dizi olarak oluşturur.

2. `salary = np.arange(40000, 76000, 5000)`: 40.000'den başlayarak, 5.000'lik artışlarla 76.000'e kadar olan maaşları bir dizi olarak oluşturur.

3. `salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83545]`: İkinci bir maaş veri seti, belirli değerlerle manuel olarak oluşturulmuştur.

4. `plt.figure(figsize=(8, 5))`: Grafiğin boyutunu 8x5 birim olarak ayarlar.

5. `plt.plot(age, salary, "b:", label="SALARY")`: İlk maaş veri setini mavi noktalı çizgi ("b:") ile çizer ve etiketini "SALARY" olarak belirler.

6. `plt.plot(age, salary_2, "r-", label="SALARY 2")`: İkinci maaş veri setini kırmızı çizgi ("r-") ile çizer ve etiketini "SALARY 2" olarak belirler.

7. `plt.xlabel("AGE")` ve `plt.ylabel("SALARY")`: Sırasıyla x ve y eksenlerini "AGE" ve "SALARY" olarak etiketler.

8. `plt.title(label="SALARY BY AGE")`: Grafiğe "SALARY BY AGE" başlığını ekler.

9. `plt.legend()`: Çizimdeki etiketleri (legend) gösterir.

10. `plt.show()`: Çizilen grafiği gösterir.


````python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83545]
````

````python
%matplotlib inline
plt.figure(figsize = (8, 5))
plt.plot(age, salary, "b:", label = "SALARY")
plt.plot(age, salary_2, "r-", label = "SALARY 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")

plt.title(label = "SALARY BY AGE")
plt.legend()
plt.show()
````

#### Kod Çıktısı Yorumlaması

Çıktıda, yaşa göre maaş dağılımını gösteren iki farklı çizgi grafiği görülür. İlk veri seti ("SALARY"), mavi noktalı çizgiyle temsil edilirken, ikinci veri seti ("SALARY 2") kırmızı çizgiyle gösterilir. Bu grafikler, yaş arttıkça maaşın nasıl değiştiğini karşılaştırmalı olarak gösterir. Eksen etiketleri ve başlık, grafiğin içeriğini açıklar ve legend, hangi çizginin hangi veri setine ait olduğunu belirtir. Bu tür karşılaştırmalı grafikler, farklı veri setlerindeki trendleri ve örüntüleri görselleştirmede yararlıdır.

### Kod Açıklaması

Bu kod bloğu, Python'un Matplotlib kütüphanesini kullanarak yaşa bağlı maaş dağılımlarını gösteren iki farklı çizgi grafiğini oluşturmak ve görselleştirmek için tasarlanmıştır.

1. `%matplotlib inline`: Bu sihirli komut, Jupyter Notebook'ta çizimlerin doğrudan not defterlerinin içinde gösterilmesini sağlar. Bu, grafiklerin ayrı bir pencerede açılmasını önleyerek, analiz işlemini daha akıcı hale getirir.

2. `plt.figure(figsize = (8, 5))`: Grafiğin pencere boyutunu ayarlar. `figsize = (8, 5)` parametresi ile grafiğin genişliği 8 inç ve yüksekliği 5 inç olarak belirlenir, bu da okunabilir ve anlaşılır bir görsel sunar.

4. `plt.plot(age, salary_2, "r-", label = "SALARY 2")`: İkinci bir maaş dağılımını kırmızı kesikli çizgi (`"r-"`) ile çizer ve bu çizgiye "SALARY 2" etiketi atar. Bu, farklı bir maaş skalasını veya maaş setini temsil edebilir.

# Object Oriented Method (OOM)

### Kod Açıklaması

Bu kod bloğu, Matplotlib kütüphanesi kullanarak, iki farklı maaş veri setini karşılaştırmak için dikey olarak sıralanmış iki alt grafik içeren bir çizgi grafiği oluşturur. Kod, `subplots` fonksiyonu aracılığıyla daha düzenli ve esnek bir şekilde alt grafik yönetimi sağlar.

1. `fig, ax = plt.subplots(2, 1)`: Bu komut, iki satır ve bir sütun olacak şekilde düzenlenmiş iki alt grafiği içeren bir figür (`fig`) ve alt grafikler için eksenler (`ax`) oluşturur. `ax` bir dizi olarak döndürülür, her bir elemanı bir alt grafiğin eksenlerini temsil eder.

2. `ax[0].plot(age, salary, "r")`: Dizinin ilk elemanı (`ax[0]`), yaş (`age`) ve ilk maaş seti (`salary`) verilerini kullanarak kırmızı (`r`) renkli bir çizgi grafiği çizer. Bu, veri setinin yaşa bağlı maaş dağılımını temsil eder.

3. `ax[1].plot(age, salary_2, "b")`: Dizinin ikinci elemanı (`ax[1]`), yaş (`age`) ve ikinci maaş seti (`salary_2`) verilerini kullanarak mavi (`b`) renkli bir çizgi grafiği çizer. Bu, farklı bir veri setinin yaşa bağlı maaş dağılımını gösterir.

4. `plt.tight_layout()`: Bu komut, figürdeki alt grafikler arasındaki boşlukları otomatik olarak ayarlar, böylece etiketler, başlıklar ve eksen isimleri gibi ögeler üst üste binmez ve her şey net bir şekilde görülebilir hale gelir.

5. `plt.show()`: Oluşturulan çizgi grafiğini ekranda gösterir.

````python
fig, ax = plt.subplots(2, 1)

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout()
plt.show()
````

#### Kod Çıktısı Yorumlaması

Çıktı olarak, iki alt grafikten oluşan bir çizgi grafiği seti elde edilir. Her alt grafik, yaşa göre maaş dağılımını farklı bir veri seti için gösterir. İlk alt grafik kırmızı çizgi ile `salary` veri setini, ikinci alt grafik ise mavi çizgi ile `salary_2` veri setini temsil eder. `plt.tight_layout()` komutunun kullanılması sayesinde, grafikler arasında yeterli boşluk bırakılarak tüm detayların net bir şekilde görülmesi sağlanır. Bu düzenleme, iki farklı maaş veri setinin karşılaştırılmasını kolaylaştırır ve yaşa göre değişim trendlerini açıkça gösterir.

## Functional Method | OOP 

İlk kod bloğu, Matplotlib'in fonksiyonel yaklaşımını kullanır. Bu yaklaşımda, plt.subplot fonksiyonu ile doğrudan alt grafikler oluşturulur ve her bir alt grafik için plt.plot fonksiyonu çağrılarak çizimler yapılır. Bu yöntem, daha basit görselleştirmeler için hızlı ve kolay bir yol sunar.



* İkinci kod bloğu ise Matplotlib'in nesne yönelimli programlama (OOP) yaklaşımını kullanır. Bu yaklaşımda, plt.subplots fonksiyonu kullanılarak bir figür ve alt grafiklerin eksenleri (ax) bir kezde oluşturulur. Daha sonra, bu eksen objeleri üzerinden çizimler yapılır. Bu yöntem, daha karmaşık görselleştirmeler ve ince ayarlar için daha fazla kontrol ve esneklik sağlar.



Her iki yöntemin de avantajları vardır:

* Fonksiyonel yöntem, basit görselleştirmeler için hızlı ve kolay bir yol sunar.
* OOP yöntemi, daha karmaşık görselleştirmelerde ve çizimlerde daha iyi bir kontrol ve esneklik sağlar, ayrıca kodun yeniden kullanılabilirliğini ve okunabilirliğini artırır.
* Sonuç olarak, iki yöntem de benzer görseller üretse de, kullanım senaryonuza ve ihtiyaçlarınıza bağlı olarak hangi yöntemin tercih edileceği değişebilir.

````python
fig, ax = plt.subplots(3,2)

ax[0][0].plot(age, salary, "r")
ax[0][1].plot(age, salary_2, "b")

ax[1][0].plot(age, salary_2, "b")
ax[1][1].plot(age, salary, "r")

ax[2][0].plot(age, salary_2, "b")
ax[2][1].plot(age, salary, "r")

plt.tight_layout()
plt.show()
````

Doğrusal regresyon çizgisi, iki değişken arasındaki ilişkiyi daha net anlamamıza yardımcı olur.

1. `fig, ax = plt.subplots()`: Bir figür ve bir eksen oluşturur. Bu yaklaşım, grafiğin çeşitli özelliklerini daha kolay kontrol etmemizi sağlar.

2. `ax.scatter(tips.total_bill, tips.tip)`: `scatter` metodu ile `tips` veri setinden `total_bill` (toplam fatura) ve `tip` (bahşiş) sütunları kullanılarak saçılım grafiği çizilir. Her bir nokta, bir restoran müşterisinin ödediği toplam fatura ve bıraktığı bahşiş miktarını temsil eder.

4. `m, b = np.polyfit(x, y, 1)`: `np.polyfit` fonksiyonu, `x` ve `y` üzerinde basit bir doğrusal regresyon (1. dereceden polinom) hesaplar ve modelin eğimi (`m`) ve y-intercept'i (`b`) olmak üzere iki parametre döndürür. Bu, veri setine en uygun doğruyu bulmamızı sağlar.

5. `ax.plot(x, m*x + b, c="r")`: Hesaplanan eğim ve y-intercept kullanılarak, saçılım grafiği üzerine kırmızı (`c="r"`) renkte bir doğrusal regresyon çizgisi çizilir. Bu çizgi, `tips` veri setindeki `total_bill` ve `tip` arasındaki genel ilişkiyi gösterir.

6. `plt.show()`: Oluşturulan saçılım grafiği ve üzerine eklenen doğrusal regresyon çizgisini gösterir.

https://www.cuemath.com/geometry/y-mx-b/

### İKİNCİ BİR METHOD

**Veri Çerçevesini Seçme:**

- **numeric_tips = tips.select_dtypes(include=["float64", "int64"])**: Bu satır, tips adlı veri çerçevesinden sadece float64 ve int64 türlerindeki (yani sayısal türdeki) sütunları seçer ve bu sütunlardan oluşan yeni bir veri çerçevesi oluşturur. Bu işlem, analizde yalnızca sayısal verilerin kullanılmasını istediğinizde yararlıdır.

**Gruplama ve Toplama:**

- **numeric_tips.groupby(tips["day"]).sum()**: Bu satır, numeric_tips veri çerçevesini tips["day"] sütununa göre gruplar ve her grubun sayısal değerlerinin toplamını hesaplar. Bu, day sütunundaki her bir gün için, sayısal sütunların toplamlarını almak demektir. Örneğin, eğer tips veri çerçevesinde total_bill, tip gibi sayısal sütunlar varsa ve day sütunu haftanın günlerini içeriyorsa (örneğin, Perşembe, Cuma, vs.), bu kod her bir gün için toplam hesabı ve bahşiş miktarını verecektir.

````python
numeric_tips = tips.select_dtypes(include = ["float64", "int64"])

numeric_tips.groupby(tips["day"]).sum()
````

- **["total_bill"].sum()**: Her bir grup için (day bazında gruplandırılmış), total_bill sütununun toplamını hesaplar. Bu, her gün için toplam hesap miktarını verir.

**İndeks Alma:**

- **.index**: Bu toplamın indekslerini alır. Bu durumda, indeksler day sütunundaki benzersiz değerlerdir (yani günler). groupby işlemi sonucunda elde edilen toplam, aslında bir pandas Series nesnesidir ve bu nesnenin index özelliği, gruplama işlemi için kullanılan benzersiz değerleri (bu örnekte günleri) içerir.

Sonuç olarak, day değişkeni, tips DataFrame'indeki farklı günleri içeren bir indeks listesi olur. Bu liste, hangi günlerin veri setinde yer aldığını ve dolayısıyla analizde hangi günlerin dikkate alındığını gösterir.

1. `plt.figure(figsize=(10, 8))`: Pasta grafiğinin çizileceği figürün boyutunu belirler. Bu, figürün genişliğinin 10 inç ve yüksekliğinin 8 inç olacağı anlamına gelir, böylece grafik detayları daha net görülebilir.

2. `sex_counts = tips['sex'].value_counts()`: `tips` veri setindeki `sex` sütunundaki değerlerin sayısını hesaplar. Bu, veri setindeki erkek ve kadın müşteri sayısını belirler ve sonuç, cinsiyet kategorilerine göre sıralanmış bir pandas Serisi olarak döndürülür.
